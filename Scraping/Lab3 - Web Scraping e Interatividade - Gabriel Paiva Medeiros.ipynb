## Entrega: Gráficos Interativos

Use **Plotly** (pode ser plotly express, se preferir)


**Questão 1**: Para cada gênero qual o nome dos  deputados(as) que tiveram maior quantidade de presença no plenário? (dica: scatter plot com mouse over)

```python
new_df = df[df.genero == 'M']
fig = px.scatter(new_df, y="presenca_plenario", hover_data=["nome"])
fig.update_layout(
  title = {
      'text' : '<b>' + 'Presença Mulheres'+ '</b>',
      'x': 0.5,
      'xanchor': 'center'
  },
)
fig.show()
```

```python
new_df = df[df.genero == 'H']
fig = px.scatter(new_df, y="presenca_plenario", hover_data=["nome"])
fig.update_layout(
  title = {
      'text' : '<b>' + 'Presença Homens'+ '</b>',
      'x': 0.5,
      'xanchor': 'center'
  },
)
fig.show()
```

**Questão 2**: Como se comporta o total de gastos parlamentar e de gabinete em cada mês do ano de 2022: (dica: use um gráfico de barra (stacked ou não))

```python
df.columns
```

```python
new_df = df[[
    'gasto_jan_par', 'gasto_fev_par', 'gasto_mar_par',
    'gasto_abr_par', 'gasto_maio_par', 'gasto_junho_par', 'gasto_jul_par',
    'gasto_agosto_par', 'gasto_set_par', 'gasto_out_par', 'gasto_nov_par'
]]
fig = px.histogram(x=new_df.columns, y=list(new_df.sum()))
fig.update_layout(
  title = {
      'text' : '<b>' + 'Gasto Total Mensal - Parlamento'+ '</b>',
      'x': 0.5,
      'xanchor': 'center'
  },
  xaxis_title="Meses",
  yaxis_title="Gasto Total",
)
fig.show()
```

```python
new_df = df[[
    'gasto_jan_gab', 'gasto_fev_gab', 'gasto_mar_gab', 'gasto_abr_gab',
    'gasto_maio_gab', 'gasto_junho_gab', 'gasto_jul_gab',
    'gasto_agosto_gab', 'gasto_set_gab', 'gasto_out_gab', 'gasto_nov_gab'
]]

fig = px.histogram(x=new_df.columns, y=list(new_df.sum()))
fig.update_layout(
  title = {
      'text' : '<b>' + 'Gasto Total Mensal - Gabinete'+ '</b>',
      'x': 0.5,
      'xanchor': 'center'
  },
  xaxis_title="Meses",
  yaxis_title="Gasto Total",
)
fig.show()
```
