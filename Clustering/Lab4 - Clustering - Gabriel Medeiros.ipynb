## Imports.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import plotly as py
import plotly.graph_objs as go
from sklearn.cluster import KMeans
```

## Carregando o **Dataset**

```python
# Dataset:
# https://drive.google.com/file/d/1Ez7TUtxvBZeRdIUvCT6KPqI9qirmkLSA/view?usp=sharing

df = pd.read_csv('clientes.csv')
df.head()
```

```python
df = df.drop(['ID'], axis=1)
df = df.drop(['Genero'], axis=1)
df
```

## Primeiro Agrupamento (Cálculo + Plot)

### Definindo a entrada do algoritmo (X)

```python
new_df = df.drop(['Idade'], axis=1)
plt.scatter(df['Renda'], df['Pontuacao Consumo'])
```

```python
X = new_df.values
```

### Cálculo

```python
algorithm = KMeans(n_clusters = 5)
algorithm.fit(X)
labels = algorithm.labels_
centroids = algorithm.cluster_centers_
```

### Plot

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='plasma')
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, alpha=0.5)
```
